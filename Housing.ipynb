```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dftr = pd.read_csv('train.csv') 
dftt = pd.read_csv('test.csv')

dftt
```

```python
# droping feautures that are missing more than 50%
col = dftr.columns.values

dropit = []

for i in col:
    if(dftr[i].isnull().sum()>700):
        dropit.append(i)

dftr_drop = dftr.drop(dropit, axis=1)
dftt_drop = dftt.drop(dropit, axis=1)


dftr_drop.shape
```

```python
dftr_drop.corr().style.background_gradient(cmap = 'BuPu')
```

```python
dftr_drop.corr().SalePrice
```

```python
features = []

for i,j in dftr_drop.corr().SalePrice.iteritems():
    if(j>0.1 and i != 'SalePrice'):
        features.append(i)

features
```

```python
for i in features:
    dftr_drop[i].fillna(dftr_drop[i].mean(), inplace=True)
    dftt_drop[i].fillna(dftt_drop[i].mean(), inplace=True)
```

```python
for i in features:
    print(dftr_drop[i].isnull().sum())
    print(dftt_drop[i].isnull().sum())
```

```python
cat_features = dftr_drop.select_dtypes(include = 'object').columns.values
cat_features
```

```python
for i in cat_features:
    dftr_drop[i].fillna(dftr_drop[i].mode()[0] , inplace=True)
    dftt_drop[i].fillna(dftr_drop[i].mode()[0] , inplace=True)
```

```python
for i in cat_features:
    print(dftr_drop[i].isnull().sum())
    print(dftt_drop[i].isnull().sum())
```

```python
for i in cat_features:
    features.append(i)
print(features)
```

```python
a = dftr_drop[features]
b = dftt_drop[features]
com_data = pd.concat((a, b), axis=0, ignore_index= True)
com_data
```

```python
com_data = pd.get_dummies(com_data, columns = cat_features)

data_cat = com_data.select_dtypes(include="object").columns.values
com_data = pd.get_dummies(com_data,columns = data_cat)

train_data_x = com_data.loc[0:1459, : ]
train_data_y = np.log(dftr['SalePrice'])
test_data = com_data.loc[1460:, : ]

print(test_data.shape)
print(train_data_x.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_data_x, train_data_y, test_size=0.15, random_state = 2021)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
# X_test = scaler.transform(test_data)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

lr = RandomForestRegressor()

lr.fit(X_train, y_train)

print(mean_squared_error(np.exp(y_test), np.exp(lr.predict(X_test)), squared = False))
print(mean_squared_error(np.exp(y_train), np.exp(lr.predict(X_train)), squared = False))
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
rf = RandomForestRegressor(random_state=2021)
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(train_data_x, train_data_y)
```

```python
from sklearn.model_selection import GridSearchCV
# Create the parameter grid based on the results of random search 
param_grid = {
    'bootstrap': [True],
    'max_depth': [80, 90, 100, 110],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200, 300, 1000]
}
# Create a based model
rf = RandomForestRegressor(2021)
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
```

```python
grid_search.fit(train_data_x, train_data_y)
```

```python
print(mean_squared_error(np.exp(y_test), np.exp(grid_search.predict(X_test)), squared = False))
print(mean_squared_error(np.exp(y_train), np.exp(grid_search.predict(X_train)), squared = False))
```

```python
print(mean_squared_error(np.exp(y_test), np.exp(rf_random.predict(X_test)), squared = False))
print(mean_squared_error(np.exp(y_train), np.exp(rf_random.predict(X_train)), squared = False))
```
